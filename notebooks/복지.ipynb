```python
import pandas as pd

df = pd.read_excel('인구십만명당문화기반시설수2024.xlsx', names=('행정구역','문화/인구', '문화기반시설수', '주민등록인구'))
df = df[1:]
df.head()
```

```python
df.info()
```

```python
df['행정구역'] = df['행정구역'].str.strip()
df[df['행정구역']=='종로구'].index
```

```python
df_1 = df[78:]
df_2 = df[:78]
```

```python
df_total = pd.concat([df_2, df_1], ignore_index=True)
df_total.head()
```
