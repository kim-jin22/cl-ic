```python
import pandas as pd, re, requests, time

df = pd.read_csv('[생활]_공원(면적포함).csv')  # 파일명 확인 후 수정

# 타겟 지역 필터
targets = ['서울특별시', '인천광역시', '경기도 성남시', '경기도 고양시', '경기도 부천시',
           '경기도 광명시', '경기도 하남시', '경기도 남양주시', '경기도 용인시',
           '경기도 김포시', '경기도 구리시', '경기도 안양시']

df_t = df[df['소재지지번주소'].apply(
    lambda x: any(str(x).startswith(t) for t in targets)
)].copy().reset_index(drop=True)

print(f'타겟 공원 수: {len(df_t)}')

# 카카오 API로 행정동 추출
KAKAO_API_KEY = 'YOUR_KAKAO_API_KEY'

def get_행정동(lat, lon):
    url = 'https://dapi.kakao.com/v2/local/geo/coord2regioncode.json'
    headers = {'Authorization': f'KakaoAK {KAKAO_API_KEY}'}
    params = {'x': lon, 'y': lat}
    try:
        res = requests.get(url, headers=headers, params=params)
        docs = res.json().get('documents', [])
        for doc in docs:
            if doc['region_type'] == 'H':
                return doc['region_3depth_name']
    except:
        pass
    return None

df_t['행정동'] = None
total = len(df_t)

for i, row in df_t.iterrows():
    dong = get_행정동(row['위도'], row['경도'])
    df_t.at[i, '행정동'] = dong
    time.sleep(0.05)
    
    if i % 200 == 0:
        print(f'{i}/{total} ({i/total*100:.1f}%)')
    if i % 500 == 0 and i > 0:
        df_t.to_csv('공원_행정동추가_중간.csv', index=False, encoding='utf-8-sig')

print('완료!')
df_t.to_csv('[생활]_공원(면적포함).csv_행정동추가.csv', index=False, encoding='utf-8-sig')
```
