```python
import requests
import pandas as pd
import time

url = "https://dapi.kakao.com/v2/local/search/address.json"
headers = {"Authorization": "KakaoAK YOUR_KAKAO_API_KEY"}

def get_address(address):
    try:
        params = {'query' : address}
        res = requests.get(url, headers=headers, params=params, timeout=5)
        a = res.json()
        sigungoo = a['documents'][0]['address']['region_2depth_name']   # 시군구
        # law_dong = a['documents'][0]['address']['region_3depth_name']   # 법정동
        # hang_dong = a['documents'][0]['address']['region_3depth_h_name']   # 행정동
        return sigungoo
        
    except Exception as e:
        print(f"주소 변환 실패: {address} / 에러: {e}")
        return pd.Series([None, None])

```

```python
import requests
import pandas as pd
import time

url = "https://dapi.kakao.com/v2/local/geo/coord2address.json"
headers = {"Authorization": "KakaoAK YOUR_KAKAO_API_KEY"}

def get_dong(xy):
    x = xy['X좌표값']
    y = xy['Y좌표값']
    try:
        params = {'x' : x, 'y' : y}
        res = requests.get(url, headers=headers, params=params, timeout=5)
        a = res.json()

        sigungoo = a['documents'][0]['address']['region_2depth_name']
        return sigungoo

    except Exception as e:
        print(f"주소 변환 실패: {xy} / 에러: {e}")
        return None
```

```python
df_merge2=df_merge[df_merge.columns[:25].to_list() + df_merge.columns[117:136].to_list()]
```

```python
df_merge2.drop(columns=(['남자', '여자']), inplace=True)
```

```python
df_merge2['합계']=df_merge2.iloc[:,4:23].sum(axis=1) + df_merge2.iloc[:,23:].sum(axis=1)
```

```python
df_merge2=df_merge2.drop(columns=(['남자','여자']))
```

```python
df_merge2['센터지수(정규화전)'] = (df_merge2['시설수']/df_merge2['합계']) *1000
```

```python
df_merge2= df_merge2.dropna(subset=['0세남자'])
```

```python
max_center = df_merge2['센터지수(정규화전)'].max()
min_center = df_merge2['센터지수(정규화전)'].min()
df_merge2['센터지수(정규화후)'] = (df_merge2['센터지수(정규화전)'] - min_center) / (max_center - min_center)
```

```python
df_merge2['가중치 적용'] = df_merge2['센터지수(정규화후)'] * 0.0804
```

```python
max_center = df_merge2['가중치 적용'].max()
min_center = df_merge2['가중치 적용'].min()
df_merge2['최종'] = ((df_merge2['가중치 적용'] - min_center) / (max_center - min_center)) * 100
```

```python
df_merge2.to_csv('final/[복지]지역아동센터_동별지수.csv')
```
