```python
# 1) 없는 동 추가
missing_dong = set(df2['읍면동명']) - set(df['행정동'])
missing_rows = df2[df2['읍면동명'].isin(missing_dong)][['시군구명', '읍면동명']].drop_duplicates()

missing_df = pd.DataFrame(columns=df.columns)
missing_df['행정동'] = missing_rows['읍면동명'].values

for col in df.columns:
    if col != '행정동':
        missing_df[col] = 0

df_final = pd.concat([df, missing_df], ignore_index=True)

# 2) 시군구 매핑표 만들기
dong_map = df2[['시군구명', '읍면동명']].drop_duplicates()

# 3) 전체 행에 대해 시군구 붙이기
df_final = df_final.merge(
    dong_map,
    left_on='행정동',
    right_on='읍면동명',
    how='left'
)

# 4) 컬럼명 정리
df_final['시군구'] = df_final['시군구명']
df_final.drop(columns=['시군구명', '읍면동명'], inplace=True)

df_final
```
