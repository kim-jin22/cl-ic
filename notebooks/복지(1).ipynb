```python
import pandas as pd

data = pd.read_csv('data/경기도_지역아동센터현황.csv', encoding='cp949')
data.head()
```

```python
df = data[data['시군명'].isin(['성남시', '고양시', '부천시', '광명시', '하남시', '남양주시', '용인시', '김포시', '구리시', '안양시'])]
```

```python
df = df[['시군명', '시설명', '소재지도로명주소', '소재지지번주소', 'WGS84위도', 'WGS84경도']]
```

```python
import requests
import pandas as pd
import time

url = "https://dapi.kakao.com/v2/local/search/address.json"
headers = {"Authorization": "KakaoAK YOUR_KAKAO_API_KEY"}
params = {"query": "경기도 고양시 덕양구 토당로 78번길 7 22 2층"}

def get_address(address):
    try:
        params = {'query' : address}
        res = requests.get(url, headers=headers, params=params, timeout=5)
        a = res.json()
        lat = a['documents'][0]['y']   # 위도
        lon = a['documents'][0]['x']   # 경도
        return pd.Series([lat, lon])
        
    except Exception as e:
        print(f"주소 변환 실패: {address} / 에러: {e}")
        return pd.Series([None, None])


df.loc[df['WGS84위도'].isna(),  ['WGS84위도', 'WGS84경도']] = df.loc[df['WGS84위도'].isna(), '소재지도로명주소'].apply(get_address).values
# print(df.loc[df['WGS84위도'].isna(), '소재지도로명주소'].apply(get_address).values)
```

```python
import requests
import pandas as pd
import time

url = "https://dapi.kakao.com/v2/local/geo/coord2address.json"
headers = {"Authorization": "KakaoAK YOUR_KAKAO_API_KEY"}

def get_dong(xy):
    x = xy['WGS84경도']
    y = xy['WGS84위도']
    try:
        params = {'x' : x, 'y' : y}
        res = requests.get(url, headers=headers, params=params, timeout=5)
        a = res.json()

        lat = a['documents'][0]['address']['region_3depth_name']
        return lat

    except Exception as e:
        print(f"주소 변환 실패: {xy} / 에러: {e}")
        return None


df['행정동'] = df[['WGS84경도', 'WGS84위도']].apply(get_dong, axis=1)
```

```python
df.to_csv('data_cleaning/경기도_지역아동센터현황_동추가.csv')
```
