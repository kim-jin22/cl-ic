```python
mid_df
```

```python
result = pd.merge(
    ele_df, mid_df,
    on='행정동',
    how='outer',
    suffixes=('_초', '_중')
)
result
```

```python
result = result.fillna(0)
```

```python
cols = ['학생수(계)_초','학생수(계)_중','학급수(계)_초','학급수(계)_중']
result[cols] = result[cols].astype(int)
result
```

```python
result['총학생수'] = result['학생수(계)_초'] + result['학생수(계)_중']
result['총학급수'] = result['학급수(계)_초'] + result['학급수(계)_중']

result['학급당학생수'] = result['총학생수'] / result['총학급수']
result
```

```python
result.to_csv('최종-인천학급당학생수.csv', index=False, encoding='utf-8-sig')
```
