```python
cities = ['성남', '고양', '부천', '광명', '하남', '남양주', '용인', '김포', '구리', '안양']

def filter_region(df, cities, col_region='Unnamed: 1', col_city='Unnamed: 2'):
    return df[
        (df[col_region].str.contains('서울|인천', na=False)) |
        (
            (df[col_region].str.contains('경기', na=False)) &
            (df[col_city].str.contains('|'.join(cities), na=False))
        )
    ]
```

```python
munhaw = [df_guck, df_gong, df_pack, df_mi, df_sang, df_munyea, df_gibang, df_munhouse, df_mungyeon, df_burock]
```

```python
munhaw = {
    'guck': df_guck,
    'gong': df_gong,
    'pack': df_pack,
    'mi': df_mi,
    'sang': df_sang,
    'munyea': df_munyea,
    'gibang': df_gibang,
    'munhouse': df_munhouse,
    'mungyeon': df_mungyeon,
    'burock': df_burock
}

munhaw_filtered = {}

for name, df in munhaw.items():
    munhaw_filtered[name + '_c'] = filter_region(df, cities)
```

```python
import requests
import pandas as pd
import time

url = "https://dapi.kakao.com/v2/local/search/address.json"
headers = {"Authorization": "KakaoAK YOUR_KAKAO_API_KEY"}

def get_address(address):
    try:
        params = {'query' : address}
        res = requests.get(url, headers=headers, params=params, timeout=5)
        a = res.json()
        law_dong = a['documents'][0]['address']['region_3depth_name']   # 법정동
        hang_dong = a['documents'][0]['address']['region_3depth_h_name']   # 행정동
        return pd.Series([law_dong, hang_dong])
        
    except Exception as e:
        print(f"주소 변환 실패: {address} / 에러: {e}")
        return pd.Series([None, None])

```

```python
for i in munhaw_filtered.keys():
    munhaw_filtered[i][['법정동', '행정동']] = munhaw_filtered[i].apply(get_address)
```

```python
for key, df in munhaw_filtered.items():
    addr_col = df.columns[3]   # 3번째 컬럼
    
    df[['법정동', '행정동']] = df[addr_col].apply(get_address)
```

```python
new_cols = ['시도', '시군구', '이름', '주소', '법정동', '행정동']

for key, df in munhaw_filtered.items():
    df.columns = new_cols
```
