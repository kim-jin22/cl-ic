```python
import time
import requests
import pandas as pd

# ✅ 1. 카카오 REST API 키 넣기 (여기!!)
REST_API_KEY = "YOUR_KAKAO_API_KEY"

# ✅ 2. 주소 → (위도, 경도, 법정동) 변환 함수
def get_addr_info(address):
    url = 'https://dapi.kakao.com/v2/local/search/address.json'
    headers = {"Authorization": f"KakaoAK {REST_API_KEY}"}
    params = {'query': address}
    
    try:
        response = requests.get(url, headers=headers, params=params)
        if response.status_code == 200:
            result = response.json()
            if result['documents']:
                addr_data = result['documents'][0]['address']
                lat = addr_data['y']  # 위도
                lon = addr_data['x']  # 경도
                b_dong = addr_data['region_3depth_name']  # 법정동
                h_dong = addr_data['region_3depth_h_name'] #행정동
                
                return lat, lon, b_dong, h_dong
        return None, None, None, None
    except:
        return None, None, None, None

# ✅ 3. CSV 불러오기 (네 파일)
df = pd.read_csv('교육 - 초등학교(인천).csv', encoding='cp949')

# ✅ 4. API 적용 (주소 컬럼 기준)
print("좌표 변환 중...")
info_list = []

for addr in df['도로명주소']:
    info = get_addr_info(addr)
    info_list.append(info)
    time.sleep(0.2)  # ⭐ API 제한 방지 (중요)

# ✅ 5. 컬럼 추가
df['위도'], df['경도'], df['법정동'], df['행정동'] = zip(*info_list)

# ✅ 6. 저장
df.to_csv('인천초등학교_법정동행정동.csv', index=False, encoding='utf-8-sig')

print("완료!")
df.head()
```
