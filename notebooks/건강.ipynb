```python
import pandas as pd
```

```python
df = pd.read_csv('data_dong/[건강]아동대상의사_동별지수.csv')
```

```python
df
```

```python
df['가중치'] = df['아동대상의사_정규화'] * 0.1275
```

```python
(df['아동대상의사_원값'] - df['아동대상의사_원값'].min()) / (df['아동대상의사_원값'].max() - df['아동대상의사_원값'].min())
```

```python
df.rename(columns={'가중치':'아동대상의사_가중치'}, inplace=True)
```

```python
df.to_csv('final/[건강]아동대상의사_동별지수.csv')
```
