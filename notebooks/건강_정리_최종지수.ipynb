```python
import requests
import pandas as pd
import time

url = "https://dapi.kakao.com/v2/local/geo/coord2regioncode.json"
headers = {"Authorization": "KakaoAK YOUR_KAKAO_API_KEY"}

def get_dong(lat, lon):
    try:
        params = {"x": lon, "y": lat}
        res = requests.get(url, headers=headers, params=params, timeout=5)
        data = res.json()

        time.sleep(0.2)

        for doc in data.get('documents', []):
            if doc['region_type'] == 'H':
                return doc['region_3depth_name']

        return None
    except:
        return None
```

```python
import pandas as pd

df_vac = pd.read_csv("[건강]예방접종_동별지수.csv")
df_ped = pd.read_excel("[건강]소아과.xlsx")

df_ped = df_ped.drop(columns=["Unnamed: 0"], errors="ignore")
df_ped = df_ped.rename(columns={"읍면동명": "행정동"})
```

```python
# 같은 시도 내에서 행정동이 하나의 시군구에만 속하는 경우만 남김
dong_map = (
    df_ped.groupby(["시도명", "행정동"])["시군구명"]
    .nunique()
    .reset_index(name="시군구개수")
)

# 유일한 동만 추출
unique_dong = dong_map[dong_map["시군구개수"] == 1][["시도명", "행정동"]]

# 유일한 동에 대한 시군구명 매핑표 만들기
unique_map = df_ped.merge(unique_dong, on=["시도명", "행정동"], how="inner")
unique_map = unique_map[["시도명", "행정동", "시군구명"]].drop_duplicates()
```

```python
df_vac_fix = df_vac.merge(
    unique_map,
    on=["시도명", "행정동"],
    how="left",
    suffixes=("", "_보정")
)

# 보정값이 있으면 그걸 쓰고, 없으면 원래 값 유지
df_vac_fix["시군구명"] = df_vac_fix["시군구명_보정"].fillna(df_vac_fix["시군구명"])
df_vac_fix = df_vac_fix.drop(columns=["시군구명_보정"])
```

```python
df_vac_use = df_vac_fix[["시도명", "시군구명", "행정동", "예방접종_가중치"]].copy()
df_ped_use = df_ped[["시도명", "시군구명", "행정동", "소아청소년과_가중치점수"]].copy()

df_merge = pd.merge(
    df_vac_use,
    df_ped_use,
    on=["시도명", "시군구명", "행정동"],
    how="left"
)
```

```python
df_merge["최종"] = (
    df_merge["예방접종_가중치"].fillna(0)
    + df_merge["소아청소년과_가중치점수"].fillna(0)
)
```
