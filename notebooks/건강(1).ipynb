```python
import pandas as pd
import requests
import time
```

```python
data_h_add = pd.read_excel('data/전국 병의원 및 약국 현황 2025.12/1.병원정보서비스(2025.12.).xlsx')
data_h_soa = pd.read_excel('data/전국 병의원 및 약국 현황 2025.12/5.의료기관별상세정보서비스_03_진료과목정보 2025.12..xlsx')
```

```python
df_h_soa = data_h_soa[['요양기관명', '진료과목코드명','과목별 전문의수', '선택진료 의사수']]
```

```python
df_h_soa
```

```python
df_h_add = data_h_add[['요양기관명','종별코드명','시군구코드명', '읍면동', '주소', '좌표(X)', '좌표(Y)']]
```

```python
df_h_add
```

```python
df_h_soa = df_h_soa[df_h_soa['진료과목코드명'] == '소아청소년과']
```

```python
df_marge = pd.merge(df_h_soa, df_h_add, on='요양기관명', how='inner')
```

```python
df_merge = df_marge[df_marge['과목별 전문의수'] != 0]
```

```python
df_merge.to_csv('data_cleaning/지역별_소아청소년과전문의수.csv', encoding='utf-8')
```

```python
df_seuol = pd.read_csv('data_cleaning/서울시_지역아동센터_ 동추가.csv')
df_incheon = pd.read_csv('data_cleaning/인천_지역아동센터_동추가.csv')
df_gi = pd.read_csv('data_cleaning/경기도_지역아동센터현황_동추가.csv')
```

```python
df_gi = df_gi.rename(columns={'WGS84위도':'Y좌표값', 'WGS84경도':'X좌표값'})
```

```python
df_gi = df_gi.drop('소재지지번주소', axis=1)
df_gi = df_gi.rename(columns={'소재지도로명주소':'주소'})
df_gi = df_gi.drop(columns=['시군명', 'Unnamed: 0'])
```

```python
df_gi.head()
```

```python
df_incheon = df_incheon.drop('군구명', axis=1)
df_incheon = df_incheon.drop('Unnamed: 0', axis=1)
```

```python
df_incheon.head()
```

```python
df_seuol = df_seuol.drop(columns=['자치구명', 'Unnamed: 0', '상세주소'])
df_seuol = df_seuol.rename(columns={'기본주소':'주소'})
```

```python
df_seuol.head()
```

```python
df_child = pd.concat([df_gi, df_seuol, df_incheon])
```

```python
df_child.to_csv('data_cleaning/통합지역아동복지센터.csv')
```
