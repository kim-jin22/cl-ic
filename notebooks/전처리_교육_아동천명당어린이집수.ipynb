```python
df = df.merge(
    df2[['시군구명', '읍면동명']],
    left_on='행정동',
    right_on='읍면동명',
    how='left'
)

df['시군구'] = df['시군구명']
df.drop(columns=['시군구명', '읍면동명'], inplace=True)

df
```

```python
df = pd.read_csv('어린이집합계3.csv', encoding='cp949')
df
```

```python
df['유아천명당시설수'] = df['어린이집수'] / df['유아인구합계'] * 1000
df
```

```python
df.to_csv('어린이집합계3.csv', index=False, encoding='utf-8-sig')
```

```python
min_val = df['유아천명당시설수'].min()
max_val = df['유아천명당시설수'].max()

df['정규화값'] = (df['유아천명당시설수'] - min_val) / (max_val - min_val)
df
```

```python
df['가중치적용값'] = df['정규화값'] * 0.0820
df
```

```python
df.to_csv('어린이집합계3.csv', index=False, encoding='utf-8-sig')
```
